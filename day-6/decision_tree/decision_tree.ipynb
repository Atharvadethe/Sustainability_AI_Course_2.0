```python
import pandas as pd
import numpy as np

data = {
    'Ripe': [1, 1, 1, 1, 0, 0],
    'Wormy': [0, 0, 1, 0, 0, 1],
    'Washed': [1, 0, 1, 0, 1, 1],
    'Eat': ['Yes', 'No', 'No', 'No', 'No', 'No']
}
df = pd.DataFrame(data)
df
```

```python
#define features and target 
X = df[['Ripe', 'Wormy', 'Washed']]
y = df['Eat']
from sklearn.tree import DecisionTreeClassifier
# Create the decision tree classifier
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
# Fit the classifier to the data
clf.fit(X, y)
ripe = int(input("Is the fruit ripe? (1 for Yes, 0 for No): "))
wormy = int(input("Is the fruit wormy? (1 for Yes, 0 for No): "))
washed = int(input("Is the fruit washed? (1 for Yes, 0 for No): "))
features = np.array([[ripe, wormy, washed]])
prediction = clf.predict(features)
print(f"The prediction is: {prediction[0]}")
# Visualizing the decision tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.title("Decision Tree for Fruit Eating Decision")
plt.show()
```
