```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', linestyle='-', color='b', label='Data Points')
plt.title('Sample Plot')
```

```python
x = np.array([5, 7, 8, 7, 17, 2, 9, 4, 11, 12, 9, 6]) #Age of the car
y = np.array([56, 78, 89, 78, 200, 20, 97, 46, 110, 150, 97, 56]) #Carbon emission
plt.scatter(x, y, label = "Honda Car")
x = np.array([6, 7, 9, 8, 16, 3, 9, 4, 11, 12, 9, 7]) #Age of the car
y = np.array([50, 70, 98, 80, 250, 20, 97, 46, 120, 140, 96, 66]) #Carbon emission
plt.scatter(x, y, label = "Maruti Suzuki Cars")
plt.title("CO2 emission by aged cars")
plt.xlabel("Age of the car")
plt.ylabel("Carbon emission in Kg CO2")
plt.legend()
plt.show()
```

```python
#different types of plots
plt.hist(x, bins=5, color='blue', alpha=0.7, label='Histogram of x')
plt.figure()  # Create a new figure for the bar chart
plt.bar(x, y, color='green', alpha=0.5, label='Bar Chart of y')
plt.figure()  # Create a new figure for the pie chart
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=140, labeldistance=1.1, colors=plt.cm.Paired.colors)
plt.title('Different Types of Plots')
plt.show()
```

```python
#draw a pie chart andapply explode effect to it for renewable energy sources and also mentuin start angle
values = [30, 20, 25, 15, 10]
labels = ['Solar', 'Wind', 'Hydro', 'Geothermal', 'Biomass']
explode = (0.1, 0, 0, 0, 0)  
myColors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']
plt.pie(values, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140, colors=myColors)
plt.title('Renewable Energy Sources Distribution')
plt.axis('equal')  
plt.show()
```

```python
import seaborn as sns

sns.set_theme()
fmri = sns.load_dataset("fmri")
sns.relplot(kind= "line", x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

#scatterplot
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
#violinplot

sns.violinplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
# pairplot
g = sns.pairplot(tips, hue='day', markers=["o", "s", "D", "X"])
g.fig.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()

# correlation heatmap (only numeric columns)
sns.heatmap(tips.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```
