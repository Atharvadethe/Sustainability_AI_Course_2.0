```python
import numpy as np
import pandas as pd
energy_Data = pd .read_csv('energy_consumption.csv')

```

```python
energy_Data.fillna(method='ffill')
```

```python
energy_Data.isna().sum()
```

PREPROCESSING 

```python
#import minmax scaler
from sklearn.preprocessing import MinMaxScaler
#initialize the scaler
scaler = MinMaxScaler()

numeric_data = energy_Data.select_dtypes(include=[np.number])
energy_Data_scaled = scaler.fit_transform(numeric_data)

# Convert the scaled data back to a DataFrame
energy_Data_scaled = pd.DataFrame(energy_Data_scaled, columns=numeric_data.columns)

print(energy_Data_scaled)
```
